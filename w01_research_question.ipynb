## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
import pandas as pd
import os
from pathlib import Path

# Find the dataset automatically
possible_paths = [
    "data/raw/content_refresh_anonymized.csv",
    "data/content_refresh_anonymized.csv",
    "content_refresh_anonymized.csv"
]

csv_path = None

for path in possible_paths:
    if os.path.exists(path):
        csv_path = path
        break

# If still not found, search the entire repository
if csv_path is None:
    for root, dirs, files in os.walk("."):
        if "content_refresh_anonymized.csv" in files:
            csv_path = os.path.join(root, "content_refresh_anonymized.csv")
            break

if csv_path is None:
    raise FileNotFoundError(
        "content_refresh_anonymized.csv not found anywhere in the repository."
    )

print("Dataset found at:", csv_path)

# Load dataset
df = pd.read_csv(csv_path)

# Display statistics
print("\nTotal rows:", len(df))
print("Total columns:", len(df.columns))
print("Average CTR:", round(df["ctr"].mean(), 4))
print("Average Search Volume:", round(df["search_volume"].mean(), 2))

df.head()
```
